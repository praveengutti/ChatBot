```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
from tkinter import *
import pyttsx3 as pp
from tkinter import messagebox
import wikipedia
import time

#setting voice for chatbot
engine=pp.init()
voices=engine.getProperty('voices')
engine.setProperty('voice',voices[1].id)
def speak(word):
    engine.say(word)
    engine.runAndWait()
bot=ChatBot('mybot')


#training chatbot with conversation
conversation = [
    "Hello",
    "Hi there!",
    "How are you?",
    "I'm doing great.",
    "That is good to hear",
    "Thank you.",
    "You're welcome."
    ,'what is your name?',
    'My name is Bot.',
    'who created you?',
    'I am created by Gaurav.',
    'where are you working now?',
    'I am working at Accenture.',
    'In which institute are you studying now?',
    'I am persuing Data Science at AlmaBetter.'
    ]

trainers=ListTrainer(bot)
trainers.train(conversation)


#setting frames
main = Tk()
main.geometry("500x730")              #set size of window
main.title("My ChatBot")

#set title of chat bot
img = PhotoImage(file='robot.png')
photoL = Label(main, image=img)
photoL.pack(pady=5)
###########################################################

#show current time and date
def start():

    text=time.strftime("%d-%m-%Y   %H:%M:%S  %p")
    label.config(text=text)
    label.after(200,start)
label=Label(main,bg="#ffffff",fg="red",font=("verdana",20,"bold"))
label.pack()
start()

############################################################
#define a function to chat with bot
def ask_from_bot ():
    query=text.get()
    answer_from_bot=bot.get_response(query)
    msgs.insert(END,'you :' + query)
    speak(query)
    msgs.insert(END, 'bot :' + str(answer_from_bot))
    speak(answer_from_bot)                                           #getting output from bot in voice
    text.delete(0,END)
    msgs.yview(END)

# create a menu bar and quit button
menubar = Menu(main)
menubar.add_command(label="Quit!", command=main.quit)
main.config(menu=menubar)
frame = Frame(main)

#set a scrollbar
sc = Scrollbar(frame)                                   #set scrollbar
msgs = Listbox(frame, width=80, height=10,yscrollcommand=sc.set)              #provide listbox
sc.pack(side=RIGHT, fill=Y)
msgs.pack(side=LEFT, fill=BOTH, pady=10)

frame.pack()
text = Entry(main, font=("Verdana", 20))  # format text
text.pack(fill=X, pady=10)

#button- ask from bot
btn = Button(main, text="Ask from bot",bg='#000000' , font=('Verdana', 20), fg='red', command = ask_from_bot)
btn.pack(expand=True,fill="both")

#binding keyboard key-(enter) with btn
def enter_function(event):
    btn.invoke()
main.bind('<Return>',enter_function)

#####################################################################################################################
#calculator in bot
val = ""
A = 0
operator = ""


def calculator():

    def btn1_clicked():                                #alloting btn in calci-0 to 9, +, -, *, /, C, =
        global val
        val = val + "1"
        data.set(val)

    def btn2_clicked():
        global val
        val = val + "2"
        data.set(val)

    def btn3_clicked():
        global val
        val = val + "3"
        data.set(val)

    def btn4_clicked():
        global val
        val = val + "4"
        data.set(val)

    def btn5_clicked():
        global val
        val = val + "5"
        data.set(val)

    def btn6_clicked():
        global val
        val = val + "6"
        data.set(val)

    def btn7_clicked():
        global val
        val = val + "7"
        data.set(val)

    def btn8_clicked():
        global val
        val = val + "8"
        data.set(val)

    def btn9_clicked():
        global val
        val = val + "9"
        data.set(val)

    def btn0_clicked():
        global val
        val = val + "0"
        data.set(val)

    def btn_plus():
        global A
        global operator
        global val
        A = int(val)
        operator= "+"
        val = val+ "+"
        data.set(val)

    def btn_minus():
        global A
        global operator
        global val
        A = int(val)
        operator = "-"
        val = val + "-"
        data.set(val)

    def btn_multiply():
        global A
        global operator
        global val
        A = int(val)
        operator = "*"
        val = val + "*"
        data.set(val)

    def btn_divide():
        global A
        global operator
        global val
        A = int(val)
        operator = "/"
        val = val + "/"
        data.set(val)

    def btn_C():
        global A
        global operator
        global val
        A=0
        val=""
        operator=""
        data.set(val)

    def result():
        global A
        global operator
        global val
        val2=val
        if operator=="+":
            x=int((val2.split("+")[1]))
            C=A+x
            data.set(C)
            val=str(C)
        elif operator=="-":
            x=int((val2.split("-")[1]))
            C=A-x
            data.set(C)
            val=str(C)
        elif operator=="*":
            x=int((val2.split("*")[1]))
            C=A*x
            data.set(C)
            val=str(C)
        elif operator=="/":
            x=int((val2.split("/")[1]))
            if x==0:
                messagebox.showerror("Error","Division by 0 not supported")
                A=""
                val=""
                data.set(val)
            else:
                C=A/x
                data.set(C)
                val=str(C)


    def on_cancel():
        top.destroy()

    top = Toplevel(main)
    top.geometry("300x450")
    top.resizable(0, 0)

    top.title("Calculator")


#menubar with back button in calci
    menubar = Menu(top)
    menubar.add_command(label="< Back", command=on_cancel)
    top.config(menu=menubar)

    data=StringVar()
    lbl=Label(top,text="Label",anchor=SE,font=('verdana',20),textvariable=data,bg='#ffffff',fg='#000000')
    lbl.pack(expand=True,fill="both")

    btnrow1=Frame(top,bg="#000000")
    btnrow1.pack(expand=TRUE,fill="both")

    btnrow2 = Frame(top)
    btnrow2.pack(expand=TRUE, fill="both")

    btnrow3 = Frame(top)
    btnrow3.pack(expand=TRUE, fill="both")

    btnrow4 = Frame(top)
    btnrow4.pack(expand=TRUE, fill="both")


    # creating buttons in btnrow1
    btn1=Button(btnrow1,text="1",font=("verdana",22),relief=GROOVE,border=0,command=btn1_clicked)
    btn1.pack(side=LEFT,expand=True,fill="both")

    btn2 = Button(btnrow1, text="2", font=("verdana", 22),relief=GROOVE,border=0,command=btn2_clicked)
    btn2.pack(side=LEFT, expand=True, fill="both")

    btn3 = Button(btnrow1, text="3", font=("verdana", 22),relief=GROOVE,border=0,command=btn3_clicked)
    btn3.pack(side=LEFT, expand=True, fill="both")

    btna = Button(btnrow1, text="+", font=("verdana", 22),relief=GROOVE,border=0,command = btn_plus)
    btna.pack(side=LEFT, expand=True, fill="both")



    # creating buttons in btnrow2
    btn4=Button(btnrow2,text="4",font=("verdana",22),relief=GROOVE,border=0,command=btn4_clicked)
    btn4.pack(side=LEFT,expand=True,fill="both")

    btn5 = Button(btnrow2, text="5", font=("verdana", 22),relief=GROOVE,border=0,command=btn5_clicked)
    btn5.pack(side=LEFT, expand=True, fill="both")

    btn6 = Button(btnrow2, text="6", font=("verdana", 22),relief=GROOVE,border=0,command=btn6_clicked)
    btn6.pack(side=LEFT, expand=True, fill="both")

    btns = Button(btnrow2, text="-", font=("verdana", 22),relief=GROOVE,border=0,command=btn_minus)
    btns.pack(side=LEFT, expand=True, fill="both")




    # creating buttons in btnrow3
    btn7=Button(btnrow3,text="7",font=("verdana",22),relief=GROOVE,border=0,command=btn7_clicked)
    btn7.pack(side=LEFT,expand=True,fill="both")

    btn8 = Button(btnrow3, text="8", font=("verdana", 22),relief=GROOVE,border=0,command=btn8_clicked)
    btn8.pack(side=LEFT, expand=True, fill="both")

    btn9 = Button(btnrow3, text="9", font=("verdana", 22),relief=GROOVE,border=0,command=btn9_clicked)
    btn9.pack(side=LEFT, expand=True, fill="both")

    btnm = Button(btnrow3, text="*", font=("verdana", 22),relief=GROOVE,border=0,command=btn_multiply)
    btnm.pack(side=LEFT, expand=True, fill="both")




    # creating buttons in btnrow4
    btnz=Button(btnrow4,text="C",font=("verdana",22),relief=GROOVE,border=0,command=btn_C)
    btnz.pack(side=LEFT,expand=True,fill="both")

    btn0 = Button(btnrow4, text="0", font=("verdana", 22),relief=GROOVE,border=0,command=btn0_clicked)
    btn0.pack(side=LEFT, expand=True, fill="both")

    btni = Button(btnrow4, text="=", font=("verdana", 22),relief=GROOVE,border=0,command=result)
    btni.pack(side=LEFT, expand=True, fill="both")

    btnd= Button(btnrow4, text="/", font=("verdana", 22),relief=GROOVE,border=0,command=btn_divide)
    btnd.pack(side=LEFT, expand=True, fill="both")

    top.mainloop()


#bind calculator button with Control key + c
btnc = Button(main, text="Calculator",fg="red", command=calculator)
btnc.place(x=0,y=0)

def calculator_function(event):
    btnc.invoke()
main.bind('<Control-Key-c>',calculator_function)
###################################################################################################################

#adding wikipedia search in bot
def wiki():
    top = Toplevel(main)
    top.geometry('700x760')
    top.title("Wikipedia")
    top.resizable(0, 0)

#crate frame
    frame=Frame(top,bg="#ffffff")
    result=Frame(top,bg="#ffffff")
    input = Entry(frame, font=("Verdana", 20))  # format text
    input.pack(fill=X, pady=10)
    query=""
    text=Text(top,padx=55,pady=10,font=("verdana",16))

#setting hindi language button
    def language():
        wikipedia.set_lang("hi")

    btnlanghi = Button(frame, text="Hindi", bg='#000000', font=('Verdana', 20), fg='red',
                     command=language)
    btnlanghi.pack(side=LEFT,expand=True, fill='both')

#setting english language button
    def languageen():
        wikipedia.set_lang("en")

    btnlangen = Button(frame, text="English", bg='#000000', font=('Verdana', 20), fg='red',
                     command=languageen)
    btnlangen.pack(side=LEFT,expand=True, fill='both')

#searching on wikipedia
    def search():
        global query
        query=input.get()

        try:
            summary=wikipedia.page(query)
            text.insert('1.0',summary.content)
        except:

            text.insert('1.0',"No results")

#setting serch button
    def search_function(event):
        btnsearch.invoke()

    top.bind('<Return>', search_function)

    btnsearch = Button(frame, text="Search on Wikipedia", bg='#000000', font=('Verdana', 20), fg='red', command=search)
    btnsearch.pack(expand=True, fill='both')


    frame.pack()
    result.pack(fill="both",padx=20,pady=50)

    text.pack()


#provide back button
    def on_cancel():
        top.destroy()

    menubar = Menu(top)
    menubar.add_command(label="< Back", command=on_cancel)
    top.config(menu=menubar)

    top.mainloop()

#bind wikipedia button with control key + w
def w_function(event):
    btnwiki.invoke()

main.bind('<Control-Key-w>', w_function)


btnwiki = Button(main, text="Wikipedia", fg="red",command=wiki)

btnwiki.place(x=437, y=0)


####################################################################################################################
main.mainloop()
```
